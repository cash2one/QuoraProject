```python
import json
from matplotlib import pyplot as plt
%matplotlib inline

with open('userOut.json') as f:
    data = json.load(f)
plt.rcParams['figure.figsize'] = (20.0, 16.0)
```

```python
fig = plt.figure()
fig.suptitle("Normalized Histograms of Followers for Top Posters in Dataset")
for i, (user, followers) in enumerate(sorted(data.items(), key=lambda x:len(x[1]), reverse=True)[:9]):
    ax = plt.subplot2grid((3,3),(i / 3, i % 3))
    plt.title("User Post Rank #{}".format(i+1))
    plt.ylabel("Frequency")
    plt.xlabel("Follower Count")
    bins=list(range(min(followers), max(followers) + 1, 1))[:100]
    plt.hist(followers, bins=bins, normed=True)
    axes = plt.gca()
plt.show()
```

```python
post_count = [len(l) for n, l in data.items() if n != u'null']
bins=list(range(min(post_count), max(post_count) + 1, 1))[:100]
plt.hist(post_count, bins=bins)
plt.title("Distribution of User Post Counts")
plt.ylabel("Number of Users")
plt.xlabel("Number of Posts by a User")
plt.show()
```

```python
plt.hist(post_count, bins=bins)
axes = plt.gca()
axes.set_ylim([0,3500])
plt.title("Distribution of User Post Count (Zoomed)")
plt.ylabel("User Count")
plt.xlabel("Number of Posts by a User")
plt.show()
```
